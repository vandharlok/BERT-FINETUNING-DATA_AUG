```python
from openai import OpenAI
import os
import json

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

```

```python
perguntas = [
   
]
contexto = "remarcar consulta em uma clínica médica"
resposta = "Deseja remarcar a data da consulta?"

# Lista para armazenar os exemplos formatados
exemplos_formatados = []

for pergunta in perguntas:
    # Formatar o exemplo no formato JSONL esperado
    exemplo = {
        "prompt": f"{contexto}",
        "completion": f" {pergunta}"
    }
    exemplos_formatados.append(exemplo)

# Exibindo o resultado em formato JSON
for exemplo in exemplos_formatados:
    print(json.dumps(exemplo, ensure_ascii=False))
```
